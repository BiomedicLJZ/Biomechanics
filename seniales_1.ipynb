El teorema de Nyquist, tambin conocido como el criterio de Nyquist-Shannon, es uno de los teoremas ms importantes en el procesamiento de seales y la teora de la informacin. Este teorema establece un lmite fundamental en la tasa de muestreo que es necesaria para obtener una representacin precisa de una seal analgica en una seal digital.

El teorema de Nyquist establece que para poder reconstruir completamente una seal analgica a partir de su versin digitalizada, la tasa de muestreo debe ser al menos el doble de la frecuencia mxima presente en la seal original. En otras palabras, la frecuencia de muestreo debe ser mayor o igual que el doble de la frecuencia mxima de la seal que se est muestreando. Esta frecuencia mnima de muestreo se conoce como la frecuencia de Nyquist.

La razn por la cual el teorema de Nyquist es importante es porque, si la frecuencia de muestreo es demasiado baja, pueden aparecer efectos no deseados conocidos como aliasing. El aliasing ocurre cuando una seal analgica es muestreada a una frecuencia menor que la frecuencia de Nyquist, y las frecuencias ms altas de la seal se mezclan con las frecuencias ms bajas, lo que resulta en una seal digital que no representa con precisin la seal original.

Para entender el teorema de Nyquist en la prctica, consideremos un ejemplo de una seal analgica que contiene una frecuencia mxima de 1 kHz. Para evitar el aliasing y poder reconstruir esta seal digitalmente, la frecuencia de muestreo debe ser al menos de 2 kHz. En otras palabras, se necesitan al menos dos muestras por ciclo de la seal original para evitar el aliasing. Si la seal se muestrea a una frecuencia menor que 2 kHz, aparecer aliasing en la seal digital resultante, lo que dar lugar a informacin errnea.

El teorema de Nyquist tiene aplicaciones prcticas en muchos campos, incluyendo la comunicacin digital, la msica digital, la televisin y el procesamiento de imgenes. Por ejemplo, en la televisin digital, la seal de video se muestrea a una tasa de al menos 13,5 MHz para evitar el aliasing y garantizar una representacin precisa de la seal original. En la msica digital, se utiliza una frecuencia de muestreo de al menos 44,1 kHz para evitar el aliasing y garantizar una buena calidad de sonido.

En resumen, el teorema de Nyquist establece que la tasa de muestreo mnima necesaria para digitalizar una seal analgica debe ser al menos el doble de la frecuencia mxima presente en la seal original. Si la tasa de muestreo es demasiado baja, pueden aparecer efectos no deseados conocidos como aliasing, lo que dar lugar a informacin errnea en la seal digital resultante. Como se puede observar, al usar una frecuencia de muestreo de 5 Hz, la seal no se puede representar correctamente, ya que no se puede observar la frecuencia de 20 Hz.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la seal original
t = np.linspace(0, 1, 1000)
f = 30
y = np.sin(2 * np.pi * f * t)

# Definimos la frecuencia de muestreo
fs = 40

# Muestreamos la seal
n = np.arange(0, len(t))
x = np.sin(2 * np.pi * f * n / fs)

# Reconstruimos la seal
y_r = np.zeros(len(t))
for i in range(len(t)):
    y_r[i] = sum(x * np.sinc((i / fs - n)))

# Creamos la figura
fig, ax = plt.subplots(figsize=(10, 6))

# Dibujamos la seal original y la seal reconstruida
ax.plot(t, y, label='Seal original', color='blue')
ax.plot(t, y_r, label='Seal reconstruida', color='red')

# Configuramos el ttulo y las etiquetas de los ejes
ax.set_title('Ejemplo de aliasing para f=30 Hz', fontsize=18)
ax.set_xlabel('Tiempo (s)', fontsize=14)
ax.set_ylabel('Amplitud', fontsize=14)

# Mostramos la leyenda
ax.legend(fontsize=14)

# Mostramos la figura
plt.show()
```
