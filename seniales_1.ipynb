El teorema de Nyquist, también conocido como el criterio de Nyquist-Shannon, es uno de los teoremas más importantes en el procesamiento de señales y la teoría de la información. Este teorema establece un límite fundamental en la tasa de muestreo que es necesaria para obtener una representación precisa de una señal analógica en una señal digital.

El teorema de Nyquist establece que para poder reconstruir completamente una señal analógica a partir de su versión digitalizada, la tasa de muestreo debe ser al menos el doble de la frecuencia máxima presente en la señal original. En otras palabras, la frecuencia de muestreo debe ser mayor o igual que el doble de la frecuencia máxima de la señal que se está muestreando. Esta frecuencia mínima de muestreo se conoce como la frecuencia de Nyquist.

La razón por la cual el teorema de Nyquist es importante es porque, si la frecuencia de muestreo es demasiado baja, pueden aparecer efectos no deseados conocidos como aliasing. El aliasing ocurre cuando una señal analógica es muestreada a una frecuencia menor que la frecuencia de Nyquist, y las frecuencias más altas de la señal se mezclan con las frecuencias más bajas, lo que resulta en una señal digital que no representa con precisión la señal original.

Para entender el teorema de Nyquist en la práctica, consideremos un ejemplo de una señal analógica que contiene una frecuencia máxima de 1 kHz. Para evitar el aliasing y poder reconstruir esta señal digitalmente, la frecuencia de muestreo debe ser al menos de 2 kHz. En otras palabras, se necesitan al menos dos muestras por ciclo de la señal original para evitar el aliasing. Si la señal se muestrea a una frecuencia menor que 2 kHz, aparecerá aliasing en la señal digital resultante, lo que dará lugar a información errónea.

El teorema de Nyquist tiene aplicaciones prácticas en muchos campos, incluyendo la comunicación digital, la música digital, la televisión y el procesamiento de imágenes. Por ejemplo, en la televisión digital, la señal de video se muestrea a una tasa de al menos 13,5 MHz para evitar el aliasing y garantizar una representación precisa de la señal original. En la música digital, se utiliza una frecuencia de muestreo de al menos 44,1 kHz para evitar el aliasing y garantizar una buena calidad de sonido.

En resumen, el teorema de Nyquist establece que la tasa de muestreo mínima necesaria para digitalizar una señal analógica debe ser al menos el doble de la frecuencia máxima presente en la señal original. Si la tasa de muestreo es demasiado baja, pueden aparecer efectos no deseados conocidos como aliasing, lo que dará lugar a información errónea en la señal digital resultante. El teorema de Nyquist es fundamental en el procesamiento de señales y la teoría de la información, y tiene aplicaciones prácticas en muchos campos, incluyendo la comunicación digital, la música digital, la televisión y el procesamiento de imágenes.

Como se puede observar, al usar una frecuencia de muestreo de 5 Hz, la señal no se puede representar correctamente, ya que no se puede observar la frecuencia de 20 Hz.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la señal original
t = np.linspace(0, 1, 1000)
f = 30
y = np.sin(2 * np.pi * f * t)

# Definimos la frecuencia de muestreo
fs = 40

# Muestreamos la señal
n = np.arange(0, len(t))
x = np.sin(2 * np.pi * f * n / fs)

# Reconstruimos la señal
y_r = np.zeros(len(t))
for i in range(len(t)):
    y_r[i] = sum(x * np.sinc((i / fs - n)))

# Creamos la figura
fig, ax = plt.subplots(figsize=(10, 6))

# Dibujamos la señal original y la señal reconstruida
ax.plot(t, y, label='Señal original', color='blue')
ax.plot(t, y_r, label='Señal reconstruida', color='red')

# Configuramos el título y las etiquetas de los ejes
ax.set_title('Ejemplo de aliasing para f=30 Hz', fontsize=18)
ax.set_xlabel('Tiempo (s)', fontsize=14)
ax.set_ylabel('Amplitud', fontsize=14)

# Mostramos la leyenda
ax.legend(fontsize=14)

# Mostramos la figura
plt.show()
```
