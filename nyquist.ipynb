# Teorema de Nyquist

El teorema de *Nyquist*, también conocido como el criterio de __Nyquist-Shannon__, es uno de los teoremas más importantes en el procesamiento de señales y la teoría de la información. Este teorema establece un límite fundamental en la tasa de muestreo que es necesaria para obtener una representación precisa de una señal analógica en una señal digital.

El teorema de Nyquist establece que para poder reconstruir completamente una señal analógica a partir de su versión digitalizada, la tasa de muestreo debe ser al menos el doble de la frecuencia máxima presente en la señal original. En otras palabras, la frecuencia de muestreo debe ser mayor o igual que el doble de la frecuencia máxima de la señal que se está muestreando. Esta frecuencia mínima de muestreo se conoce como la frecuencia de Nyquist.

La razón por la cual el teorema de Nyquist es importante es porque, si la frecuencia de muestreo es demasiado baja, pueden aparecer efectos no deseados conocidos como aliasing. El aliasing ocurre cuando una señal analógica es muestreada a una frecuencia menor que la frecuencia de Nyquist, y las frecuencias más altas de la señal se mezclan con las frecuencias más bajas, lo que resulta en una señal digital que no representa con precisión la señal original.

Para entender el teorema de Nyquist en la práctica, consideremos un ejemplo de una señal analógica que contiene una frecuencia máxima de 1 kHz. Para evitar el aliasing y poder reconstruir esta señal digitalmente, la frecuencia de muestreo debe ser al menos de 2 kHz. En otras palabras, se necesitan al menos dos muestras por ciclo de la señal original para evitar el aliasing. Si la señal se muestrea a una frecuencia menor que 2 kHz, aparecerá aliasing en la señal digital resultante, lo que dará lugar a información errónea.

El teorema de Nyquist tiene aplicaciones prácticas en muchos campos, incluyendo la comunicación digital, la música digital, la televisión y el procesamiento de imágenes. Por ejemplo, en la televisión digital, la señal de video se muestrea a una tasa de al menos 13,5 MHz para evitar el aliasing y garantizar una representación precisa de la señal original. En la música digital, se utiliza una frecuencia de muestreo de al menos 44,1 kHz para evitar el aliasing y garantizar una buena calidad de sonido.

En resumen, el teorema de Nyquist establece que la tasa de muestreo mínima necesaria para digitalizar una señal analógica debe ser al menos el doble de la frecuencia máxima presente en la señal original. Si la tasa de muestreo es demasiado baja, pueden aparecer efectos no deseados conocidos como aliasing, lo que dará lugar a información errónea en la señal digital resultante.
```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la señal original
t = np.linspace(0, 1, 1000)
f = 30
y = np.sin(2 * np.pi * f * t)

# Definimos la frecuencia de muestreo
fs = 40

# Muestreamos la señal
n = np.arange(0, len(t))
x = np.sin(2 * np.pi * f * n / fs)

# Reconstruimos la señal
y_r = np.zeros(len(t))
for i in range(len(t)):
    y_r[i] = sum(x * np.sinc((i / fs - n)))

# Creamos la figura
fig, ax = plt.subplots(figsize=(20, 6))

# Dibujamos la señal original y la señal reconstruida
ax.plot(t, y, label='Señal original', color='blue', linewidth=1)
ax.plot(t, y_r, label='Señal reconstruida', color='red',linewidth=1)

# Configuramos el título y las etiquetas de los ejes
ax.set_title('Ejemplo de aliasing para f=30 Hz y fs de 40 Hz', fontsize=18)
ax.set_xlabel('Tiempo (s)', fontsize=14)
ax.set_ylabel('Amplitud', fontsize=14)

# Mostramos la leyenda
ax.legend(fontsize=14)

# Mostramos la figura
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the original signal
t = np.linspace(0, 1, 1000)
f = 30
y = np.sin(2 * np.pi * f * t)

# Define the sampling frequency
fs = 40

# Sample the signal
n = np.arange(0, len(t))
x = np.sin(2 * np.pi * f * n / fs)

# Reconstruct the signal
y_r = np.zeros(len(t))
for i in range(len(t)):
    y_r[i] = sum(x * np.sinc((i / fs - n)))

# Find the indices where the reconstructed signal meets the original signal
idx = np.where(np.round(y_r, 3) == np.round(y, 3))[0]

# Create the figure
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the original and reconstructed signals
ax.plot(t, y, label='Original Signal', color='blue', linewidth=0.5)
ax.plot(t, y_r, label='Reconstructed Signal', color='red')

# Plot points where the reconstructed signal meets the original signal
ax.scatter(t[idx], y[idx], color='green', s=10)

# Set the title and axis labels
ax.set_title('Example of Aliasing for f=30 Hz and fs of 40 Hz', fontsize=18)
ax.set_xlabel('Time (s)', fontsize=14)
ax.set_ylabel('Amplitude', fontsize=14)

# Show the legend
ax.legend(fontsize=8, loc='upper right')

# Show the figure
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
%matplotlib ipympl
# Define the function to generate the sine wave
def plot_sine_wave(freq, amp,T,fs):
    x = np.linspace(0, T,fs)  # Generate 1000 evenly spaced values from 0 to 10
    y = amp * np.sin(2*np.pi*freq * x)  # Compute the sine wave
    plt.cla()  # Clear the current figure
    plt.plot(x, y,'r')  # Plot the sine wave
    plt.ylim(-5, 5)  # Set the y-axis limits

# Use ipywidgets to create sliders for frequency and amplitude
freq_slider = widgets.FloatSlider(min=0, max=250, step=0.1, value=1, description='Frequency:')
amp_slider = widgets.FloatSlider(min=0, max=50, step=0.1, value=1, description='Amplitude:')

# Use interact to connect the sliders to the plot_sine_wave function
interactive_plot = interactive(plot_sine_wave, freq=freq_slider, amp=amp_slider,T=fixed(1),fs=fixed(1000))
output = interactive_plot.children[-1]
output.layout.height = '350px'

# Display the interactive plot
interactive_plot
```
