```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```python
def f(freq:int,time:int,fs:int,amp:int):
    '''Funcion que genera una senal senoidal
    freq: frecuencia de la senal
    T: duracion de la senal
    fs: frecuencia de muestreo
    amp: amplitud de la senal
    '''
    t=np.linspace(0,time,time*fs)
    y=amp*np.sin(2*np.pi*freq*t)
    return t,y
```

```python
t,s1=f(1,1,100,1)
```

```python
plt.plot(t,s1)
```

```python
t,s2=f(np.random.randint(1,10),1,100,1)
```

```python
plt.plot(t,s2)
```
